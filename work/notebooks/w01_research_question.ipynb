## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

Research question:
Which content pages should an SEO or content editor review first for refresh,
expansion, protection, pruning, or monitoring based on observable search and
content-performance signals?

Unit of analysis:
One pseudonymized content page.

Planned output:
A ranked list of pages with a priority score, suggested action, reason codes,
and a confidence indicator.

Decision improved:
The project will help a content team decide which pages should receive its
limited review time first.

Action:
An SEO specialist or content editor can inspect the highest-ranked pages and
decide whether to refresh, expand, protect, rewrite, monitor, or prune them.

Cost of a wrong call:
A false positive may waste editing time on a page that does not need attention.
A false negative may cause the team to miss a visible declining page or a
valuable content opportunity.

Why data or ML may help:
Several signals may interact in ways that are difficult to represent with one
fixed rule. A data-driven ranking may combine those signals more effectively,
but it must still be compared with a simple and transparent baseline.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

The starter dataset contains 30,000 content pages.

Among these pages, 13152 pages (43.84%) were declining while still receiving at least 100 impressions. I also found 9759 visible pages (32.53%) with at least 500 impressions, an average position between 1 and 20, and CTR below 0.5%.

In addition, 7076 pages (23.59%) were at least 180 days old while still appearing within the top 10 average positions.

These observed groups suggest that the dataset contains enough potentially valuable review candidates to investigate a ranked refresh-opportunity system over the next seven weeks.

```python
import pandas as pd

# Load the starter dataset from my public GitHub repository
data_url = (
    "https://raw.githubusercontent.com/"
    "Suhail-Ahmed7/flyrank-ml-internship-suhail/"
    "main/data/raw/content_refresh_anonymized.csv"
)

df = pd.read_csv(data_url)

# Number 1: total pages
total_pages = len(df)

# Number 2: pages declining while still having meaningful demand
declining_with_demand = (
    (df["trend_direction"] == "down")
    & (df["impressions_90d"] >= 100)
).sum()

# Number 3: visible pages with comparatively low CTR
low_ctr_visible = (
    (df["impressions_90d"] >= 500)
    & (df["avg_position"] > 0)
    & (df["avg_position"] <= 20)
    & (df["ctr"] < 0.5)
).sum()

# Number 4: older pages currently positioned on page one
page_one_age_risk = (
    (df["avg_position"] > 0)
    & (df["avg_position"] <= 10)
    & (df["content_age_days"] >= 180)
).sum()

summary = pd.DataFrame(
    {
        "Metric": [
            "Total pages",
            "Declining pages with at least 100 impressions",
            "Visible low-CTR pages",
            "Older pages positioned in the top 10",
        ],
        "Count": [
            total_pages,
            declining_with_demand,
            low_ctr_visible,
            page_one_age_risk,
        ],
    }
)

summary["Percentage of pages"] = (
    summary["Count"] / total_pages * 100
).round(2)

display(summary)
```
